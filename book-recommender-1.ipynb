```python
seen = []
double = []
for idx, row in df.iterrows():
    book = (row['title'], row['author'])
    if (book) not in seen:
        seen.append(book)
        double.append(False)
    else:
        double.append(True)
```

```python
print(len(seen))
print(len(double))
```

```python
double.count(True)
```

```python
df['double'] = double

df.head()
```

```python
mask = df['double'] == True

df = df[~mask]

print(df.shape)
```
