```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("english_books.csv")

print(df.shape)

df.head()
```

```python
df.isnull().sum()
```

```python
seen = []
double = []
for idx, row in df.iterrows():
    book = (row['title'], row['author'])
    if (book) not in seen:
        seen.append(book)
        double.append(False)
    else:
        double.append(True)
```

```python
df['double'] = double

df.double.value_counts()
```

```python
df.head()
```

```python
mask = df['double']

df = df[~mask]

df.shape
```

```python
df = df.drop(columns=['english', 'double'])

df.head()
```

```python
df.iloc[4]['descrip']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# tfidf = TfidfVectorizer(stop_words='english', min_df=0.05, max_df=0.95, ngram_range=(1,2))
vect = TfidfVectorizer(stop_words='english', min_df=0.05, max_df=0.95)

data = ["".join(s) for s in df['descrip']]

vect.fit(data)
```

```python
sparse = vect.transform(data)

dtm = pd.DataFrame(sparse.todense(), columns=vect.get_feature_names())

dtm.head()
```
