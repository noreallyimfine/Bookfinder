```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("english_books.csv")

print(df.shape)

df.head()
```

```python
df.isnull().sum()
```

```python
seen = []
double = []
for idx, row in df.iterrows():
    book = (row['title'], row['author'])
    if (book) not in seen:
        seen.append(book)
        double.append(False)
    else:
        double.append(True)
```

```python
df['double'] = double

df.double.value_counts()
```

```python
df.head()
```

```python
mask = df['double']

df = df[~mask]

df.shape
```

```python
df = df.drop(columns=['english', 'double'])

df.head()
```
