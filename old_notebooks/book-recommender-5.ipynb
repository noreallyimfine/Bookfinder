```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("english_books.csv")

print(df.shape)

df.head()
```

```python
df.isnull().sum()
```

```python
seen = []
double = []
for idx, row in df.iterrows():
    book = (row['title'], row['author'])
    if (book) not in seen:
        seen.append(book)
        double.append(False)
    else:
        double.append(True)
```

```python
df['double'] = double

df.double.value_counts()
```

```python
df.head()
```

```python
mask = df['double']

df = df[~mask]

df.shape
```

```python
df = df.drop(columns=['english', 'double'])

df.head()
```

```python
df.iloc[4]['descrip']
```

```python
import enchant
```

```python
d = enchant.Dict("en_US")
```

```python
def check_english(book, dictionary=d):
    book_list = book.split()
    
    return (dictionary.check(book_list[0]) and dictionary.check(book_list[1]) and dictionary.check(book_list[2]))
```

```python
df['english'] = df.descrip.apply(check_english)

df['english'].value_counts()
```

```python
mask = df['english']

non_english = df[~mask]

non_english.head(15)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# tfidf = TfidfVectorizer(stop_words='english', min_df=0.05, max_df=0.95, ngram_range=(1,2))
vect = TfidfVectorizer(stop_words='english', min_df=0.05, max_df=0.95)

data = ["".join(s) for s in df['descrip']]

vect.fit(data)
```

```python
sparse = vect.transform(data)

dtm = pd.DataFrame(sparse.todense(), columns=vect.get_feature_names())

dtm.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(n_neighbors=10, leaf_size=15, algorithm='auto', p=2, metric='euclidean')

nn.fit(dtm)
```

```python
new = vect.transform(test)
```

```python
new = new.todense()
```

```python
pd.DataFrame(new, columns=vect.get_feature_names())
```
